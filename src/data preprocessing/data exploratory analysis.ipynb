#### **i) discrete features**

```python
plt.figure(figsize=(15, 30))
sns.set_style('darkgrid')
plotnumber = 1

for column in discrete_features:
    if plotnumber < 14:   
        # row and column indices of the subplot are calculated 
        ax= plt.subplot(7, 2, plotnumber)
        sns.countplot(data=data, x=column, palette='flare',
                      orient='v')
        
        # Title, x label, y label
        ax.set_title(f'"{column}"', fontsize=15, fontweight='bold')
        ax.set_xlabel(column, fontsize=10)
        ax.set_ylabel('count', fontsize=10)

        # Increasing tick label sizes
        ax.tick_params(axis='x', labelsize=10)
        ax.tick_params(axis='y', labelsize=10)
        
        # Displaying bar values in percentage rounded off to two decimal digits with '%' sign
        # calculates the total no of datapoints for % 
        total = float(len(data))

        # iterates through each patch (bar)
        for p in ax.patches:

            # get_height()--> height of the current bar
            height = p.get_height()
            # positions the text at the center of each bar 
            ax.text(p.get_x() + p.get_width() / 2., height + 5,
                    f'{height / total * 100:.1f}%', ha="center", fontsize=10, fontweight= 'bold')

    plotnumber += 1

plt.subplots_adjust(hspace=2, wspace=2)
plt.tight_layout()
plt.show()
```

#### Insights:
- One common insight from all the above graphs seems to be that the "3" category of Performance Rating dominates in all the categorical input features.
- This represents the previous insights as similar to that of univariate analysis and when comparing the categories of each input feature, most of the employees in the dominating category have moderate performance rating of 3.
- For instance, in the **Employees education level**, the dominating category is 3rd level. Among them, most of the employees have shown moderate performance rating than the rest of the categories.
- However, the trend is different in few features.
- Eventhough the **employees have work environment satisfaction of 4th level, they show less performance rating** compared to the other categories. This goes in a similar way to that of employees job satisfaction.
